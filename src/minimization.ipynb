```python
import torch
import torch.optim as optim
import numpy as np
     
```

```python
# function to be minimized
def myfun(a): 
  return (3*torch.sigmoid(a) - 2)**2
```

```python
x = torch.tensor([-2], dtype=torch.float, requires_grad=True) #init
optimizer = optim.SGD([x], lr=0.1) #optimizer
     
```

```python
epoch_n = 210
print_n = 10
epoch_per_print= np.floor(epoch_n/print_n)
```

```python
for epoch in range(epoch_n):
    optimizer.zero_grad()
    y = myfun(x)
    y.backward(retain_graph=True)
    optimizer.step()

    if (epoch+1) % epoch_per_print == 0:
      print(f'Epoch [{epoch+1}/{epoch_n}], Loss: {y.item():.4f}, x: {x.item():.4f}')
```
