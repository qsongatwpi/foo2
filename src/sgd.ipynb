```python
import torch
import torch.optim as optim
import numpy as np
```

```python
# function to be minimized
myfun = lambda a: (3 * torch.sigmoid(a) - 2) ** 2
# graph of the function
import matplotlib.pyplot as plt
x = torch.linspace(-10, 10, 100)
y = myfun(x)
plt.plot(x.detach().numpy(), y.detach().numpy())
plt.xlabel('x')
plt.ylabel('myfun(x)')
plt.title('Graph of myfun(x)')
plt.grid()
plt.show()
```

```python
x = torch.tensor([5], dtype=torch.float, requires_grad=True) #init
optimizer = optim.SGD([x], lr=0.1) #optimizer
epoch_n = 2100
print_n = 10
epoch_per_print= np.floor(epoch_n/print_n)

for epoch in range(epoch_n):
    optimizer.zero_grad()
    y = myfun(x)
    y.backward(retain_graph=True)
    optimizer.step()

    if (epoch+1) % epoch_per_print == 0:
      print(f'Epoch [{epoch+1}/{epoch_n}], Loss: {y.item():.4f}, x: {x.item():.4f}')
```
