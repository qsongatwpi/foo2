```python
# learn about class
class MyClass:
    def __init__(self, value=5, 
                 function=None):
        self.value = value
        self.function = function

    def display(self):
        print(f"The value is: {self.value}")
    def apply_function(self, x):
        if self.function is not None:
            return self.function(x)
        else:
            raise ValueError("No function provided.")
# create an instance of MyClass
my_instance = MyClass(value=10, function=lambda x: x**2)
my_instance.display()  # Output: The value is: 10
result = my_instance.apply_function(3)
print(f"The result of applying the function to 3 is: {result}")  # Output: The result of applying the function to 3 is: 9

```
